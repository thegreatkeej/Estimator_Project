STEP 2: TRANSFORM

```python
# Using Google Search API to grab addresses

locations = Test_Locate['vendor']

headquarters = []

for location in locations:
    params = {
      "q": f"where is {location} headquarters address",
      "location": "United States",
      "hl": "en",
      "gl": "us",
      "api_key": homeDepot_API
    }
    search = GoogleSearch(params)
    results = search.get_dict()
    location_results = results["organic_results"]
    
    try:
        locate = location_results[1]['snippet_highlighted_words']
        headquarters.append(locate)
        
    except (KeyError):
        locate = 'online_only'
        headquarters.append(locate)
                                     
    print(locate)
```

```python
# Created location table

location_table = Test_Locate.assign(location=headquarters)
location_table
```

STEP 4: PREPROCESS DATA

```python
# Collect all data

# path1 = "C:\\Users\\kijah\\Documents\\Data_Bootcamp\\Estimator_Project\\all_vendor_table.csv"
# vendorTable = pd.read_csv(path1)

path2 = "C:\\Users\\kijah\\Documents\\Data_Bootcamp\\Estimator_Project\\data.csv"
data = pd.read_csv(path2)

data.dtypes
```

```python
data.head()
```

```python
data['zip'] = (np.where(data['Location'].str.contains(','), data['Location'].str.split(',').str[-1], data['Location']))
#     df.drop(columns =['VEND'], inplace = True)
data['zip'] = data['zip'].str.strip()
data['zip'] = (np.where(data['zip'].str.len()>5, "other", data['zip']))
```

```python
# # Example of how to look at rows with a particular value

# data[data['zip']=='other']
```

```python
# Create Regional columns

region = []
for i in data['zip']:
    if i.startswith("0"):
        region.append('northeast')
    elif i.startswith("1"):
        region.append('northeast')
    elif i.startswith("2"):
        region.append('east')
    elif i.startswith("3"):
        region.append('southeast')
    elif i.startswith("4"):
        region.append('midwest')
    elif i.startswith("5"):
        region.append('northplaines')
    elif i.startswith("6"):
        region.append('central')
    elif i.startswith("7"):
        region.append('south')
    elif i.startswith("8"):
        region.append('mountains')
    elif i.startswith("9"):
        region.append('west')
    else:
        region.append('other')
        
        
data['region'] = region
```
