```python
# Dependencies
from config import homeDepot_API
from serpapi import GoogleSearch
import pandas as pd
import numpy as np
import os
os.getcwd()
```

STEP 1: EXTRACT

```python
# Created Working 'Material Description by grabbing names from Home Depot'

common_name = []

ids = ['100400409', '100400406', '100137321', '100400405', 
       '202241090', '100144234', '202077145', '202077147', 
       '202288508', '202304774', '202304649', '202304645', 
       '202019377', '202316275', '202019375', '202316276']

for i in ids:

    params = {
      "engine": "home_depot_product",
      "product_id": i,
      "api_key": homeDepot_API
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    products = results['product_results']
    
    common_name.append(products['title'])
```

```python
# Order and establish index_name

index_name = []

index_name.append(common_name[3])
index_name.append(common_name[1])
index_name.append(common_name[0])
index_name.append(common_name[5])
index_name.append(common_name[8])
index_name.append(common_name[4])
index_name.append(common_name[2])
index_name.append(common_name[6])
index_name.append(common_name[7])
index_name.append(common_name[12])
index_name.append(common_name[13])
index_name.append(common_name[14])
index_name.append(common_name[15])
index_name.append(common_name[9])
index_name.append(common_name[10])
index_name.append(common_name[11])

index_name
```

```python
def searcher(Q, T, L, P):
    params = {
      "engine": "google",
      "q": Q,
      "api_key": homeDepot_API
    }
    search = GoogleSearch(params)
    results = search.get_dict()
    products = results["organic_results"]
   
    for p in products[0:10]:
        try:
            print(p['title'])
            T.append(p['title'])
            L.append(p['link'])
            print(p['rich_snippet']['top']['detected_extensions']['price'])
            P.append(p['rich_snippet']['top']['detected_extensions']['price'])
                        
            print('-------------------------------------------')
        except (KeyError):
            P.append('NaN')
            print ('NaN')
```

STEP 2: TRANSFORM

```python
# Grab raw data
for i in range(16):
    T_i = []
    L_i = []
    P_i = []
    searcher(index_name[i], T_i, L_i, P_i)
    
# Create a data frame out of each item category
    df_i = pd.DataFrame(data={'Manufactures Description': T_i, 'link': L_i, 
                              'Material Price': P_i, 'VENDOR': L_i})
    df_i['Vendor Location'] = (np.where(df_i['VENDOR'].str.contains('.'),
                  df_i['VENDOR'].str.split('.').str[1], df_i['VENDOR']))
    df_i.drop(columns =['VENDOR'], inplace = True)

# Change data type of price to numeric and drop NaN's
    df_i['Material Price'] = df_i['Material Price'].apply(pd.to_numeric, args=('coerce',))
    df_i.dropna(inplace=True)

# Drop price values below $0.98
    df_i.drop(df_i[df_i['Material Price'] <= .98].index, inplace = True)

# Cut off all vendors with only 1 value count
    df_i = df_i.loc[df_i.duplicated(subset='Vendor Location', keep=False), :]
```
