```python
def searcher(q, pg):

    params = {
      "engine": "home_depot",
      "q": q,
      "page": pg,
        "api_key": homeDepot_API
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    products = results["products"]
    return    products
```
