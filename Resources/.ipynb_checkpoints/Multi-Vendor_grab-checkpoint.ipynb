```python
# Dependencies
from config import homeDepot_API
from serpapi import GoogleSearch
import pandas as pd
import numpy as np
import os
os.getcwd()
```

STEP 1: EXTRACT

```python
# Created Working 'Material Description by grabbing names from Home Depot'

items = []

ids = ['100400409', '100400406', '100137321', '100400405', 
       '202241090', '100144234', '202077145', '202077147', 
       '202288508', '202304774', '202304649', '202304645', 
       '202019377', '202316275', '202019375', '202316276']

for i in ids:

    params = {
      "engine": "home_depot_product",
      "product_id": i,
      "api_key": homeDepot_API
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    products = results['product_results']
    
    items.append(products['title'])
```

```python
# Created Lists that will later become columns

title = []
links = []
prices = []
```

```python
# Created function to querry
def searcher(querrys):
    params = {
      "engine": "google",
      "q": querrys,
      "api_key": homeDepot_API
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    products = results["organic_results"]
    
    for p in products[0:10]:
        try:
            print(p['title'])
            title.append(p['title'])
            links.append(p['link'])
            print(p['rich_snippet']['top']['detected_extensions']['price'])
            prices.append(p['rich_snippet']['top']['detected_extensions']['price'])
                        
            print('-------------------------------------------')
        except (KeyError):
            prices.append('NaN')
            print ('NaN')
```

```python
# The search can only look for two items at a time. This script appends items to lists and prints what has been appended

zero = ['1/2 in. x 10 ft. Electric Metallic Tube (EMT) Conduit', '3/4 in. x 10 ft. Electric Metallic Tube (EMT) Conduit']
for Q in zero:
    searcher(Q)
        
one = ['1 in. x 10 ft. Electric Metallic Tube (EMT) Conduit', '1/2 in. emt set-screw connectors (5-pack)']
for Q in one:
    searcher(Q)
        
two = ['3/4 in. emt set-screw connectors (5-pack)', '1 in. emt set-screw connectors']
for Q in two:
    searcher(Q)
        
three = ['1/2 in. emt set-screw coupling (5-pack)', '3/4 in. emt set-screw coupling (5-pack)']
for Q in three:
    searcher(Q)
        
four = ['1 in. emt set-screw coupling', '500 ft. #4 black stranded cu simpull thhn wire']
for Q in four:
    searcher(Q)
        
five = ['500 ft. #6 black stranded cu simpull thhn wire', '500 ft. #10 black stranded cu simpull thhn wire']
for Q in five:
    searcher(Q)
        
six = ['250 ft. 14/2 romex', '250 ft. 14/3 romex']
for Q in six:
    searcher(Q)
        
seven = ['250 ft. 12/2 romex', '250 ft. 12/3 romex']
for Q in seven:
    searcher(Q)
```

STEP 2: TRANSFORM

```python
# Grab raw data
raw_df = pd.DataFrame(data={'Manufactures Description': title, 'link': links, 'Material Price': prices, 'VENDOR': links})

# Create 'Vendor Location' column
raw_df['Vendor Location'] = (np.where(raw_df['VENDOR'].str.contains('.'),
                  raw_df['VENDOR'].str.split('.').str[1], raw_df['VENDOR']))
raw_df.drop(columns =['VENDOR'], inplace = True)

# Change data type of price to numeric and drop NaN's
raw_df['Material Price'] = raw_df['Material Price'].apply(pd.to_numeric, args=('coerce',))
raw_df.dropna(inplace=True)

# Drop price values below $0.98
raw_df.drop(raw_df[raw_df['Material Price'] <= .98].index, inplace = True)

# Cut off all vendors with only 1 value count
raw_df = raw_df.loc[raw_df.duplicated(subset='Vendor Location', keep=False), :]

# Send a CSV and Excell copy to my computer
csvFile = "C:\\Users\\kijah\\Documents\\Data_Bootcamp\\Estimator_Project\\raw_df.csv"
XLFile = "C:\\Users\\kijah\\Documents\\Data_Bootcamp\\Estimator_Project\\raw_df.xlsx"

raw_df.to_csv(csvFile, index=False)
raw_df.to_excel(XLFile, index=False)

raw_df
```
