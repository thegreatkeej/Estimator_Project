```python
# Create Working 'Material Description by grabbing names from Home Depot'

items = []

ids = ['100400409', '100400406', '100137321', '100400405', 
       '202241090', '100144234', '202077145', '202077147', 
       '202288508', '202304774', '202304649', '202304645', 
       '202019377', '202316275', '202019375', '202316276']

for i in ids:

    params = {
      "engine": "home_depot_product",
      "product_id": i,
      "api_key": homeDepot_API
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    products = results['product_results']
    
    items.append(products['title'])
```

```python
title = []
links = []
prices = []
vendors = []
Test = []
```

```python
def searcher(querrys):
    params = {
      "engine": "google",
      "q": querrys,
      "api_key": homeDepot_API
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    products = results["organic_results"]

    for p in products[0:10]:
        try:
            print(p['title'])
            title.append(p['title'])
            links.append(p['link'])
            print(p['rich_snippet']['top']['detected_extensions']['price'])
            prices.append(p['rich_snippet']['top']['detected_extensions']['price'])
            vendors.append(p['link'])
            
            print('-------------------------------------------')
        except (KeyError):
            print ('NaN')
```

```python
zero = ['1/2 in. x 10 ft. Electric Metallic Tube (EMT) Conduit', '3/4 in. x 10 ft. Electric Metallic Tube (EMT) Conduit']
for Q in zero:
    searcher(Q)
        
one = ['1 in. x 10 ft. Electric Metallic Tube (EMT) Conduit', '1/2 in. emt set-screw connectors (5-pack)']
for Q in one:
    searcher(Q)
        
two = ['3/4 in. emt set-screw connectors (5-pack)', '1 in. emt set-screw connectors']
for Q in two:
    searcher(Q)
        
three = ['1/2 in. emt set-screw coupling (5-pack)', '3/4 in. emt set-screw coupling (5-pack)']
for Q in three:
    searcher(Q)
        
four = ['1 in. emt set-screw coupling', '500 ft. #4 black stranded cu simpull thhn wire']
for Q in four:
    searcher(Q)
        
five = ['500 ft. #6 black stranded cu simpull thhn wire', '500 ft. #10 black stranded cu simpull thhn wire']
for Q in five:
    searcher(Q)
        
six = ['250 ft. 14/2 romex', '250 ft. 14/3 romex']
for Q in six:
    searcher(Q)
        
seven = ['250 ft. 12/2 romex', '250 ft. 12/3 romex']
for Q in seven:
    searcher(Q)
```
