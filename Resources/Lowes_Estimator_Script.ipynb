```python
from config import homeDepot_API
from serpapi import GoogleSearch
import pandas as pd
import numpy as np
import os

os.getcwd()
```

STEP 1: EXTRACT

```python
# Create Working 'Material Description by grabbing names from Home Depot'

items = []

ids = ['100400409', '100400406', '100137321', '100400405', 
       '202241090', '100144234', '202077145', '202077147', 
       '202288508', '202304774', '202304649', '202304645', 
       '202019377', '202316275', '202019375', '202316276']

for i in ids:

    params = {
      "engine": "home_depot_product",
      "product_id": i,
      "api_key": homeDepot_API
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    products = results['product_results']
    
    items.append(products['title'])
```

```python
title = []
links = []
prices = []
```

```python
def searcher(querrys):
    params = {
      "engine": "google",
      "q": querrys,
      "api_key": homeDepot_API
    }

    search = GoogleSearch(params)
    results = search.get_dict()
    products = results["organic_results"]

    for p in products[0:3]:
        try:
            print(p['title'])
            title.append(p['title'])
            links.append(p['link'])
            print(p['rich_snippet']['top']['detected_extensions']['price'])
            prices.append(p['rich_snippet']['top']['detected_extensions']['price'])
                        
            print('-------------------------------------------')
        except (KeyError):
            prices.append('NaN')
            print ('NaN')
```

```python
zero = ['1/2-in x 10-ft Metal Emt Conduit - Lowes', '3/4-in x 10-ft Metal Emt Conduit - Lowes']
for Q in zero:
    searcher(Q)
        
one = ['1-in x 10-ft Metal Emt Conduit - Lowes', 'Sigma ProConnex  1/2-in Zinc-plated Steel Set Screw Connector Conduit Fittings (5-Pack) - Lowes']
for Q in one:
    searcher(Q)
        
two = ['Sigma ProConnex  3/4-in Zinc-plated Steel Set Screw Connector Conduit Fittings (5-Pack) - Lowes', 'Sigma ProConnex  1-in Zinc-plated Steel Set Screw Connector Conduit Fittings - Lowes']
for Q in two:
    searcher(Q)
        
three = ['Sigma ProConnex  1/2-in Zinc-plated Steel Set Screw Coupling Conduit Fittings (5-Pack) - Lowes', 'Sigma ProConnex  3/4-in Zinc-plated Steel Set Screw Coupling Conduit Fittings (5-Pack) - Lowes']
for Q in three:
    searcher(Q)
        
four = ['Sigma ProConnex  1-in Zinc-plated Steel Set Screw Coupling Conduit Fittings - Lowes', 'https://www.lowes.com/pd/Southwire-SIMpull-500-ft-4-AWG-Stranded-Blue-Copper-THHN-Wire-By-the-Roll/1001078048']
for Q in four:
    searcher(Q)
        
five = ['https://www.lowes.com/pd/Southwire-SIMpull-500-ft-6-AWG-Stranded-Black-Copper-THHN-Wire-By-the-Roll/50101714', 'https://www.lowes.com/pd/Southwire-500-ft-10-AWG-Stranded-Black-Copper-THHN-Wire-By-the-Roll/50163633']
for Q in five:
    searcher(Q)
        
six = ['Southwire  Romex SIMpull 250-ft 14/2 Solid Non-Metallic Wire (By-the-Roll) - Lowes', 'Southwire  Romex SIMpull 250-ft 14/3 Solid Non-Metallic Wire (By-the-Roll) - Lowes']
for Q in six:
    searcher(Q)
        
seven = ['Southwire  Romex SIMpull 250-ft 12/2 Solid Non-Metallic Wire (By-the-Roll) - Lowes', 'Southwire  Romex SIMpull 250-ft 12/3 Solid Non-Metallic Wire (By-the-Roll) - Lowes']
for Q in seven:
    searcher(Q)
```

STEP 2: TRANSFORM

```python
# Grab raw data
LowesRaw_df = pd.DataFrame(data={'Material Description': title, 'link': links, 'Material Price': prices, 'VENDOR': links})

# Create 'Vendor Location' column
LowesRaw_df['Vendor Location'] = (np.where(LowesRaw_df['VENDOR'].str.contains('.'),
                  LowesRaw_df['VENDOR'].str.split('.').str[1], LowesRaw_df['VENDOR']))
LowesRaw_df.drop(columns =['VENDOR'], inplace = True)

# Change data type of price to numeric and drop NaN's
LowesRaw_df['Material Price'] = LowesRaw_df['Material Price'].apply(pd.to_numeric, args=('coerce',))
LowesRaw_df.dropna(inplace=True)

LowesRaw_df
```

```python
LowesRaw_df.drop_duplicates(inplace=True, ignore_index=True)
LowesRaw_df
```
